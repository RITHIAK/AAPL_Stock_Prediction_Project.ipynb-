```python
# ============================================================
# STEP 1: INSTALL & IMPORT LIBRARIES
# ============================================================
!pip install yfinance pandas numpy matplotlib seaborn scikit-learn -q

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

print("✅ STEP 1 COMPLETE: All libraries installed and imported!")

# ============================================================
# STEP 2: DOWNLOAD & EXPLORE DATA
# ============================================================
print("\n" + "="*50)
print("STEP 2: DOWNLOADING APPLE STOCK DATA...")
print("="*50)

ticker = "AAPL"
stock = yf.download(ticker, start="2024-01-01", end="2026-06-29", progress=False)

# Flatten MultiIndex columns (fixes the big error!)
stock.columns = [' '.join(col).strip() if isinstance(col, tuple) else col for col in stock.columns]
stock = stock.rename(columns={
    'Close AAPL': 'Close',
    'Open AAPL': 'Open',
    'High AAPL': 'High',
    'Low AAPL': 'Low',
    'Volume AAPL': 'Volume'
})

print("First 5 rows:")
print(stock.head())
print("\nData shape:", stock.shape)
print("\nBasic statistics:")
print(stock.describe())

print("\n✅ STEP 2 COMPLETE: Data downloaded and explored!")

# ============================================================
# STEP 3: DATA CLEANING & VISUALIZATION
# ============================================================
print("\n" + "="*50)
print("STEP 3: CLEANING & VISUALIZING DATA...")
print("="*50)

# Check missing values
print("Missing values:", stock.isnull().sum().sum())
stock = stock.fillna(method='ffill')

# Create 4 charts
plt.figure(figsize=(15, 10))

# Chart 1: Closing Price
plt.subplot(2, 2, 1)
plt.plot(stock.index, stock['Close'], color='blue', linewidth=2)
plt.title('AAPL Closing Price Over Time', fontsize=12, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True, alpha=0.3)

# Chart 2: Volume
plt.subplot(2, 2, 2)
plt.bar(stock.index, stock['Volume'], color='orange', alpha=0.7)
plt.title('AAPL Trading Volume', fontsize=12, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Volume')

# Chart 3: High vs Low
plt.subplot(2, 2, 3)
plt.plot(stock.index, stock['High'], color='green', label='High', alpha=0.7)
plt.plot(stock.index, stock['Low'], color='red', label='Low', alpha=0.7)
plt.title('AAPL High vs Low Prices', fontsize=12, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True, alpha=0.3)

# Chart 4: Daily Range
stock['Daily_Range'] = stock['High'] - stock['Low']
plt.subplot(2, 2, 4)
plt.plot(stock.index, stock['Daily_Range'], color='purple', alpha=0.7)
plt.title('AAPL Daily Price Volatility', fontsize=12, fontweight='bold')
plt.xlabel('Date')
plt.ylabel('Price Range (USD)')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n✅ STEP 3 COMPLETE: Data cleaned and visualized!")

# ============================================================
# STEP 4: FEATURE ENGINEERING & MODEL BUILDING
# ============================================================
print("\n" + "="*50)
print("STEP 4: BUILDING PREDICTION MODEL...")
print("="*50)

data = stock.copy()

# Create features
data['MA_5'] = data['Close'].rolling(window=5).mean()
data['MA_20'] = data['Close'].rolling(window=20).mean()
data['Price_Change'] = data['Close'].pct_change()
data['Price_Lag1'] = data['Close'].shift(1)
data['Price_Lag2'] = data['Close'].shift(2)
data['Volume_MA5'] = data['Volume'].rolling(window=5).mean()

# Target: Tomorrow's closing price
data['Target'] = data['Close'].shift(-1)

# Drop NaN rows
data = data.dropna()

print("Features created. Sample:")
print(data[['Close', 'MA_5', 'MA_20', 'Price_Change', 'Price_Lag1', 'Target']].head())

# Prepare for ML
features = ['Open', 'High', 'Low', 'Volume', 'MA_5', 'MA_20',
            'Price_Change', 'Price_Lag1', 'Price_Lag2', 'Volume_MA5']
X = data[features]
y = data['Target']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

print(f"\nTraining samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")

# Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\n=== MODEL PERFORMANCE ===")
print(f"RMSE: ${rmse:.2f}")
print(f"R2 Score: {r2:.4f} ({r2*100:.2f}%)")

print("\n✅ STEP 4 COMPLETE: Model trained and evaluated!")

# ============================================================
# STEP 5: FINAL VISUALIZATION & CONCLUSIONS
# ============================================================
print("\n" + "="*50)
print("STEP 5: FINAL CHARTS & CONCLUSIONS...")
print("="*50)

# Plot Actual vs Predicted
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(y_test.values, label='Actual Price', color='blue', linewidth=2)
plt.plot(y_pred, label='Predicted Price', color='red', linestyle='--', linewidth=2)
plt.title('Actual vs Predicted Stock Price', fontsize=12, fontweight='bold')
plt.xlabel('Test Period')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.scatter(y_test.values, y_pred, alpha=0.6, color='green')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.title('Prediction Accuracy Scatter', fontsize=12, fontweight='bold')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Feature Importance
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': np.abs(model.coef_)
}).sort_values('Importance', ascending=False)

print("\n=== FEATURE IMPORTANCE ===")
print(feature_importance)

plt.figure(figsize=(10, 5))
plt.barh(feature_importance['Feature'], feature_importance['Importance'], color='skyblue')
plt.title('Feature Importance', fontsize=12, fontweight='bold')
plt.xlabel('Importance Score')
plt.gca().invert_yaxis()
plt.grid(True, alpha=0.3)
plt.show()

# FINAL CONCLUSIONS - SAFE PRINT (no f-strings with logic)
close_vals = stock['Close'].values
start_price = float(close_vals[0])
end_price = float(close_vals[-1])
trend = 'upward' if end_price > start_price else 'downward'

daily_range_vals = stock['Daily_Range'].values
avg_vol = float(np.mean(daily_range_vals))

top_feat = str(feature_importance.iloc[0]['Feature'])
top_imp = float(feature_importance.iloc[0]['Importance'])

print("\n" + "="*60)
print("PROJECT CONCLUSIONS & FINDINGS")
print("="*60)
print("")
print("DATASET: Apple (AAPL) Stock Prices (Jan 2024 - Jun 2026)")
print("TOTAL RECORDS: " + str(len(stock)) + " trading days")
print("MODEL USED: Linear Regression")
print("MODEL ACCURACY: R2 Score = %.4f (%.2f%%)" % (r2, r2 * 100))
print("")
print("KEY FINDINGS:")
print("1. Stock showed %s trend from $%.2f to $%.2f" % (trend, start_price, end_price))
print("2. Most important predictor: %s (Importance: %.4f)" % (top_feat, top_imp))
print("3. Average daily volatility: $%.2f" % avg_vol)
print("4. Model RMSE: $%.2f (lower is better)" % rmse)
print("")
print("BUSINESS INSIGHT:")
print("- Moving averages help smooth out price noise")
print("- Past prices are strong predictors of future prices")
print("- Volume spikes often precede major price movements")
print("")
print("LIMITATIONS:")
print("- Stock markets are unpredictable; model is for educational purposes")
print("- External factors (news, earnings) not included")
print("- Past performance doesn't guarantee future results")
print("="*60)

print("\n🎉🎉🎉 ALL STEPS COMPLETE! PROJECT IS READY! 🎉🎉🎉")
```
